```python
import numpy as np
#pyplot
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2, 10)
y = x**2
print(x)
```

```python
plt.figure()
plt.plot(x, y, label = 'quadratique')
plt.plot(x, x**3, label= 'cubique')
plt.title('fiqure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, c='red')
plt.subplot(2, 1, 2)
plt.plot(x, y, c='blue')
```

```python
plt.subplot(2,2,1)
plt.plot(x, y)
plt.title('graph 1')
plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x), label='sinus')
plt.plot(x, np.cos(x), label='cosinus')
plt.legend()
```
