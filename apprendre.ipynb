```python
classeur = {
    "positif": [],
    "negatif": [],
}
```

```python
def trier(classeur, nombre):
    if nombre < 0:
        classeur["negatif"] = nombre
    elif nombre > 0:
        classeur["positif"] = nombre
    return classeur
```

```python
trier(classeur, 7)
```

```python
trier(classeur, -7)
```

```python
from skimage import data, color
import matplotlib.pyplot as plt

face = data.astronaut()

face_gray = color.rgb2gray(face)

plt.imshow(face_gray, cmap=plt.cm.gray)
plt.show()

print(type(face_gray))
print(face_gray.shape)

h = face_gray.shape[0]
w = face_gray.shape[1]
print(h)
print(w)
zoom_face = face_gray[h//4 : -h//4, w//4 : -w//4]
zoom_face[zoom_face > 100] = 255

plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```

```python
face_gray = face_gray[::2, ::2]
plt.imshow(face_gray, cmap=plt.cm.gray)
plt.show()
```
