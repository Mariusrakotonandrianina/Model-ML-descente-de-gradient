```python
classeur = {
    "positif": [],
    "negatif": [],
}
```

```python
def trier(classeur, nombre):
    if nombre < 0:
        classeur["negatif"] = nombre
    elif nombre > 0:
        classeur["positif"] = nombre
    return classeur
```

```python
trier(classeur, 7)
```

```python
trier(classeur, -7)
```
